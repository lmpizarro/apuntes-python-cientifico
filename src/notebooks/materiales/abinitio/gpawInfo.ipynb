## Cálculo sobre elementos simples: Al, Cu, Sn

El primer elemento tratado, es el Aluminio (Al), le sigue el Cobre (Cu) y luego el Estaño (Sn). Como elementos compuestos se calcularon propiedades para  Cu3Al y el Cu3Sn. Todos con la estructura cristalográfica fcc.
 

Para el Aluminio se hicieron 3 versiones del cálculo. El problema era llegar a los valores de referencia que los autores publican como valores de referencia. El parámetro de red tiene que ser aproximadamente a = 4.04 Angstrom.

Las 2 primeras corridas configuradas con el funcional LDA, dieron como resultado 3,98 Angstrom, la diferencia entre ellas está en la cantidad de puntos k, utilizadas en una y otra. No se obtienen diferencias significativas. 

La tercera versión del cálculo para el Aluminio se hizo con el funcional PBE, con este seteo se llega al valor de a = 4,042 Angstrom.

Algo similar se hizo con el Cu, primero se hace el cálculo usando el funcional LDA y no se llega al valor de referencia de la literatura, pero cuando se usa el funcional PBE, se obtiene el valor de equilibrio del parámetro de red a = 3,64 Angstrom, que se aproxima mejor a los valores de referencia de la literatura.

Para el estaño fcc, SnFcc, se usó el funcional LDA y se obtuvo un parámetro de red  de equilibrio a = 4.904 Angstrom. Resta calcularlo usando el funcional PBE.

## Cálculo sobre materiales compuestos

Para los materiales Cu3Al y Cu3Sn, el funcional utilizado es LDA, es por eso que hay que recalcularlos para obtener resultados comparables con los valores conseguidos por PBE para el Cu y el Al.

# Al FCC xc = 'LDA'

```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline

xAlFcc = np.array([3.5, 3.6, 3.85, 3.95,  3.97,  3.98, 3.99, 4.00, 4.01, 4.04, 4.05, 4.25, 4.5])
yAlFcc = np.array([-13.262957, -14.739886, -16.544188, -16.719516, -16.727042, -16.727917, -16.726978, -16.724018, -16.719438, -16.695562, -16.684553, -16.193775, -15.144421])

for x,y in zip(xAlFcc, yAlFcc):
    print 'a: {0:.3f}  E: {1:.3f}.'.format(x,y)  

zAlFcc = np.polyfit(xAlFcc, yAlFcc, 4)

pAlFcc = np.poly1d(zAlFcc)

xp = np.linspace(3.45, 4.55, 100)
plt.plot(xAlFcc, yAlFcc, '.', xp, pAlFcc(xp), '-');
```

```python
deriv = np.gradient(pAlFcc(xp))
plt.plot(xp,deriv)
plt.grid()
coeff = np.polyfit(xp, deriv, 3)
raicesAlFcc = np.roots(coeff)
aMinAlFcc = 3.98099523
eMinAlFcc = pAlFcc (aMinAlFcc)
print 'AlFcc a: {0:.6f} E: {1:.6f}'.format(aMinAlFcc, eMinAlFcc)
raicesAlFcc
```

# Script para cálculo fcc 2 átomos 

```python
from ase import Atom, Atoms
from gpaw import GPAW
from gpaw.test import equal

from ase.lattice.spacegroup import crystal

a = 3.775

bulk = Atoms([Atom('Sn', (0, 0, 0)),
              Atom('Cu', (0, a*0.5, a*0.5)),
              Atom('Cu', (a*0.5, a*0.5, 0)),
              Atom('Cu', (a*0.5,0 ,a*0.5))    ],
              cell=(a, a, a),
              pbc=1)

n = 8
calc = GPAW(gpts=(2*n, 2*n, 2*n),
            nbands=1*32,
            kpts=(16, 16, 16),
            convergence={'eigenstates': 2.3e-2},
            xc='LDA')


bulk.set_calculator(calc)
e2 = bulk.get_potential_energy()
niter2 = calc.get_number_of_iterations()

print e2
```
