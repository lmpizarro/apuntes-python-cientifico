## Programación Orientada a Objetos 

Hasta ahora se evitó hablar de programación orientada a objetos, pero en toda
esta presentación, usamos objetos, propiedades de  objetos y métodos de objetos.

La programación orientada a objetos está basada en clases, existe una colección de objetos
que interactuan (más o menos) entre ellos. Hasta ahora vimos la idea de progamación orientada
a funciones y rutinas. En la programación orientada a objetos, los objetos pueden recibir mensajes,
procesar datos y enviar mensajes a otros objetos.

En la imagen de las frutas, la clase primaria es Fruit, luego una subclase puede ser Manzana, otra subclase, Manzana_Red_Delicious, luego un objeto es una manzana particular, red delicious,  que la encontramos dentro de un objeto cajón.

* Encapsulado. 
   * de propiedades, color, peso, origen, fecha de cosecha.
   * de métodos, definir_la_manzana, _dame_color, _dame_peso_  
* Abstracción de datos. No tenemos una manzana, es un modelo de manzana.
* Polimorfismo. la clase fruta tiene el método _dame_peso que aplica a manzana, peras, etc
* Herencia. herencia de propiedades y métodos, el hijo agrega nuevas porpiedades y métodos.
  ejemplo: _tengo_carozo: retorna true o false

Nosotros nos vamos a dedicar a los 2 primeros, encapsulado y abstracción de datos.

Propiedades y métodos definen la clase. El valor particular de las propiedades, y los métodos específicos definen a una instancia de una clase, el objeto.

### Ejemplo 1. Una clase muy simple.

```python
# -*- coding: utf-8 -*-
#!/usr/bin/python

#
# Definición de la clase
#
class Test (object):
    # constructor de la clase
    # no siempre necesario, pero 
    # recomendable su definición
    def __init__(self, x, s):
        self.x = x #propiedad x
        self.s = s # propiedad s
        
    def metodo(self):
        print ("Esto es una función: ", self.x, self.s)

        
if __name__ == "__main__":
    # Declaración de los objetos
    # como instancias particulares
    # de la clase.
    t1 = Test(1,2)
    t2 = Test ("string", 3)
    t1.metodo()
    t2.metodo()
```

### Algo un poco más técnico

```python
class circuitRLC (object):
    def __init__ (self, R,L,C, t):
        self.R = R
        self.L = L
        self.C = C
        self.resultado =[]
        self.t = t
        
    def simulacion (self, param):
        self.param = param
        print ("Acá va la parte de la simulación parámetro: ", param)
        self.resultado = [i for i in range (self.param*self.t)]
        
    def getResultado(self):
        return (self.resultado)

if __name__ == "__main__":
    circuit1 = circuitRLC (1000, 1e-6, 10e-5,20)
    circuit2 = circuitRLC (10000, 1e-7, 20e-5, 20)
    
    circuit1.simulacion(1)
    circuit2.simulacion(2)
    
    print ("circuit 1", circuit1.getResultado())
    print ("circuit 2", circuit2.getResultado())
```

```python
class Cone ():
    def __init__(self,d0,de,L):
        "Create a cone"
        self.a0 = d0/2
        self.ae = de/2
        self.L = L
 
    def __del__(self):
        pass
 
    def radius(self,z):
        return self.ae + (self.a0-self.ae)*z/self.L
 
    def radiusp(self,z):
        # esto tambíen es un comentario
        "derivative of the radius at z"
        return (self.a0-self.ae)/self.L

# para escribir código en la misma línea usar ;
c1 = Cone(0.1,0.2,1.5); print (c1.radius(0.5))
c2 = Cone(0.2,0.2,1.5); print (c2.radius(0.25))

# defino una lista
x = []
for k in range (1,10):
    # agrego conos a la lista
    x.append (Cone (0.1*k,0.2,1.5))

# inicio un variable
k = 1
# accedo a cada uno de los conos de la lista x
for cone in x:
    # saco una cuenta e imprimo el resultado
    print (cone.radius(0.1*k))
    k = k + 1
```

## Métodos estáticos

Es un método que está escrito dentro de la clase pero que no usa la clase.

```python
class Pizza1(object):
    def __init__ (self, x, y):
        self.x = x
        self.y = y
    
    def mix_ingredients(self):
        return self.x + self.y
 
    def cook(self):
        return self.mix_ingredients(self.cheese, self.vegetables)

p = Pizza1(2,3)
print (p.mix_ingredients())

#
# Esto va a producir un error
# no puedo hacer un uso directo 
# del método sin instanciar un objeto
#
Pizza1.mix_ingredients()
```

```python
class Pizza(object):
    def __init__ (self, x, y):
        self.x = x
        self.y = y
        
    @staticmethod
    def mix_ingredients(x, y):
        return x + y
 
    def cook(self):
        return self.mix_ingredients(self.cheese, self.vegetables)
```

```python
Pizza.mix_ingredients(2,1)
```
